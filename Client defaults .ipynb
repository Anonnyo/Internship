```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("desktop\credits.csv")

df.head()
```

```python
df.info()
```

```python
dfn = pd.get_dummies(df, columns=['EDUCATION','MARRIAGE'])

dfn.head()
```

```python
final = dfn.drop(['ID'], axis = 'columns')
final
```

```python
x = final.drop('default.payment.next.month',axis = 'columns')
scaler = StandardScaler()
x = pd.DataFrame(scaler.fit_transform(x), columns=x.columns)
```

```python
y = final['default.payment.next.month']
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=1000)

model = LinearRegression()
model.fit(x_train, y_train)
model.score(x_test,y_test)
```
