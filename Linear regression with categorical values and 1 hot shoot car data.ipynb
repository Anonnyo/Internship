```python
#importig and reading csv
import pandas as pd
df = pd.read_csv("desktop\carprices.csv")
df
```

```python
# using dummies
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
# concatanate with original df

merged = pd.concat([df,dummies], axis='columns')
merged
```

```python
#dropping Car Models and 1 dummy
#for linear regression dropping dummies is not necessary 
#but for good parctice to avoid dummy variable traps

final = merged.drop(['Car Model','Mercedez Benz C class'], axis = 'columns')
final
```

```python
# calling regression and making an object
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# seperating x

x = final.drop('Sell Price($)',axis = 'columns')
x
```

```python
y = final['Sell Price($)']
y
```

```python
# training the model
model.fit(x,y)
```

```python
# prediction for the Audi A5
model.predict([[30000,3,1,0]])
```

```python
# prediction for the BMW X5
model.predict([[30000,3,0,1]])
```

```python
# prediction for the Mercedez Benz C class
model.predict([[30000,3,0,0]])
```

```python
# using score method to see how accurate the model is 
s = model.score(x,y)
print(s*100 , "%")
```
