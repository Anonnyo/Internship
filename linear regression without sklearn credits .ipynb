```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('desktop/credits.csv')
df.head()
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
x = df.drop('default.payment.next.month',axis = 1)

scaler = StandardScaler()
x = pd.DataFrame(scaler.fit_transform(x), columns=x.columns)
x
```

```python
xs = (x['LIMIT_BAL']+x['AGE']+x['BILL_AMT1']+x['BILL_AMT2']+x['BILL_AMT3']
     +x['BILL_AMT4']+x['BILL_AMT5']+x['BILL_AMT6']+x['PAY_AMT1']+x['PAY_AMT2']
     +x['PAY_AMT3']+x['PAY_AMT4']+x['PAY_AMT5']+x['PAY_AMT6']+x['SEX']+x['EDUCATION']
     +x['MARRIAGE']+x['PAY_0']+x['PAY_2']+x['PAY_3']+x['PAY_4']+x['PAY_5']+x['PAY_6'])
xs
```

```python
y = df['default.payment.next.month']

y
```

```python
plt.scatter(xs,y)
```

```python
xm = xs.mean()
xm
```

```python
df['difx'] = xm - xs
```

```python
df['difx_sq'] = df.difx**2
#df

ssx = df.difx_sq.sum()
ssx
```

```python
ym = y.mean()

df['dify'] = ym - y
df['dify_sq'] = df.dify**2
df
```

```python
ssy = df.dify_sq.sum()
ssy
```

```python
# for the slope

sp = ssy/ssx
sp
```

```python
b = ym - sp*xm
b
```

```python
def predict(value):
    predict = sp*value + b
    return predict
```

```python
predict(7)
```
