```python
#importig and reading csv
import pandas as pd
df = pd.read_csv("desktop/UCI_Credit_Card.csv")
df.info()
```

```python
# using dummies
dummies1 = pd.get_dummies(df['EDUCATION'])
dummies2 = pd.get_dummies(df['MARRIAGE'])
```

```python
merged = pd.concat([df,dummies1,dummies2], axis='columns')
merged
```

```python
final = merged.drop(['EDUCATION','MARRIAGE','ID'], axis = 'columns')
final
```

```python
x = final.drop('default.payment.next.month',axis = 'columns')
x
```

```python
y = final['default.payment.next.month']
y
```

```python
# calling regression and making an object
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# training the model
model.fit(x,y)
```

```python
#sample data

new = x.sample(500, random_state = 100)
new
```

```python
# prediction
model.predict(new)
```

```python
model.score(x,y)
```
